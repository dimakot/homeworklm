# Создать BaseCharacter, BaseInActionCharacter BaeFunkoPop, BaseCosplayer,BaseHuman
# Составить из них и из Mixin,,able логику наследований
# так, чтобы можно было минимум 6-7 Mxin'ов (созданных, а не у каждого класса)
# С помощью этих интерфейсов (мексинов) и наследования создать:
#   Shrek, PussInBoots, Donkey, JackHorner
# На каждого должен быть и персонаж, и BaeFunkoPop, и BaseCosplayer
# Итого должно быть:
# BaseCharacter
#   Shrek, PussInBoots, Donkey, JackHorner
# BaseInActionCharacter
# BaseCharacter -> BaseInActionCharacter
# BaseFunkoPop
# BaseCharacter -> BaseFunkoPop
#   Shrek -> ShrekFunkoPop, PussInBoots -> PussInBootsFunkoPop, Donkey -> DonkeyFunkoPop, JackHorner -> JackHornerFunkoPop
# BaseCosplayer
# BaseCharacter -> BaseCosplayer <- BaseHuman
#   Shrek -> ShrekCosplayer, PussInBoots -> PussInBootsCosplayer, Donkey -> DonkeyCosplayer, JackHorner -> JackHornerCosplayer

## Все классы в одном блоке

```python
class MixinSpeakable:
    def speak(self):
        return f"{self.name} говорит: \"{self.catchphrase}\""


class MixinActionable:
    def perform_action(self):
        return f"{self.name} делает: {self.signature_move}"

    def stomp_foot(self):
        return f"{self.name} топнул"

class MixinCollectible:
    is_collectible = True

    def collect(self):
        return f"Фигурка {self.name} в коллекции терь"

    def display(self):
        return f"{self.name} теперь пылится на полочке"


class MixinPoseable:
    def pose(self):
        return f"{self.name} принимает позу"

    def strike_pose(self):
        return f"{self.name} выделывается"


class MixinCostumeWearable:
    def wear_costume(self):
        return f"Костюм надет на {self.name}"

    def take_photo(self):
        return f"Сделан кадр перса {self.name}"


class MixinAnimated:
    is_animated = True

    def animate(self):
        return f"{self.name} анимируется на экране"


class MixinFunny:
    def make_laugh(self):
        return f"{self.name} шутит: {self.joke}"


class BaseHuman(MixinSpeakable):
    is_human = True

    def breathe(self):
        return "Чел делает глубокий вдох"


class BaseCharacter(MixinSpeakable, MixinAnimated, MixinFunny):
    pass


class BaseInActionCharacter(BaseCharacter, MixinActionable):
    action_mode = True
    signature_move = "атакует"


class BaseFunkoPop(BaseCharacter, MixinCollectible, MixinPoseable):
    is_vinyl_figure = True


class BaseCosplayer(BaseCharacter, BaseHuman, MixinCostumeWearable, MixinPoseable):
    costume_quality = "профессиональный"
    props_count = 5


class Shrek(BaseCharacter):
    name = "Шрек"
    catchphrase = "Ты идешь прямиком к тому, чтобы получить подзатыльник"
    joke = "Осел, ты имеешь право молчать. Но у тебя нет такой способности."


class PussInBoots(BaseCharacter):
    name = "Кот в сапогах"
    catchphrase = "Я не люблю утро"
    joke = "Я слишком сексуален для своих сапог!"


class Donkey(BaseCharacter):
    name = "Осёл"
    catchphrase = "Мне нравится этот валун. Это классный валун"
    joke = "Мы уже приехали? А сейчас?"


class JackHorner(BaseCharacter):
    name = "Джек Хорнер"
    is_villain = True
    catchphrase = "О, ну и где ты шлялся, придурок?!"
    joke = "Я волновался, что выйду голым, но моя одежда тоже выросла!"


class ShrekInAction(Shrek, BaseInActionCharacter):
    quest = "Спасти принцессу Фиону"
    signature_move = "Орет как резанный"


class PussInBootsInAction(PussInBoots, BaseInActionCharacter):
    quest = "Помочь Шреку"
    signature_move = "Глазища делать огромные чтоб все писялись от милоты"


class DonkeyInAction(Donkey, BaseInActionCharacter):
    quest = "Быть верным другом Шрека"
    signature_move = "Отвлекает врагов разговорами"


class JackHornerInAction(JackHorner, BaseInActionCharacter):
    quest = "Собрать какую-то фигню"
    signature_move = "колдун дофига чота кастует"


class ShrekFunkoPop(Shrek, BaseFunkoPop):
    rarity = "Супер редкий"
    pose_style = "стоя с поднятой рукой"


class PussInBootsFunkoPop(PussInBoots, BaseFunkoPop):
    rarity = "Легендарный"
    pose_style = "стоя с поднятым мечом"


class DonkeyFunkoPop(Donkey, BaseFunkoPop):
    rarity = "Редкий"
    pose_style = "стоя с открытым ртом"


class JackHornerFunkoPop(JackHorner, BaseFunkoPop):
    rarity = "Редкий"
    pose_style = "стоя с поднятой рукой"


class ShrekCosplayer(Shrek, BaseCosplayer):
    human_name = "Костя Шепелев"
    props_count = 3
    makeup_required = True
    green_bodysuit = True


class PussInBootsCosplayer(PussInBoots, BaseCosplayer):
    human_name = "Уважаемый Баранов Дмитрий"
    props_count = 4
    requires_sword = True
    hat_with_feather = True


class DonkeyCosplayer(Donkey, BaseCosplayer):
    human_name = "Максим Кутимский"
    props_count = 2
    donkey_ears = True
    tail_attached = True


class JackHornerCosplayer(JackHorner, BaseCosplayer):
    human_name = "Артем"
    props_count = 5
    requires_magical_props = True
    bag_of_artifacts = True
    
print("Работает короче")
```

```python
shrek = Shrek()
puss = PussInBoots()
donkey = Donkey()
jack = JackHorner()
shrek_action = ShrekInAction()
puss_action = PussInBootsInAction()
donkey_action = DonkeyInAction()
jack_action = JackHornerInAction()
shrek_funko = ShrekFunkoPop()
puss_funko = PussInBootsFunkoPop()
donkey_funko = DonkeyFunkoPop()
jack_funko = JackHornerFunkoPop()
shrek_cos = ShrekCosplayer()
puss_cos = PussInBootsCosplayer()
donkey_cos = DonkeyCosplayer()
jack_cos = JackHornerCosplayer()

print("Ну и тут")
```
