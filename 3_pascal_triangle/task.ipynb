# Выведите треугольник паскаля 
# типо вводишь циферку (до 9) и он тебе треугольник паскаля делает

```python
N, pascal_triangle = 9, []

for i in range(N):
    if i == 0:
        row = [1]
    else:
        row = [1]
        for j in range(len(pascal_triangle[-1]) - 1):
            row.append(pascal_triangle[-1][j] + pascal_triangle[-1][j + 1])
        row.append(1)
    pascal_triangle.append(row)
    
max_width = len(' '.join(f'{num:3}' for num in pascal_triangle[-1]))

for z in pascal_triangle:
    print(' '.join(f'{num:3}' for num in z).center(max_width))
```
