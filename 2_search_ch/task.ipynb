# дан регион в [-1000, 1000] 
# пользователь загадал какое-то целое число

# задача программы - замучать пользователя вопросами в виде:
#   взять середину региона и спросить больше или меньше выбранного числа 
#   и так в цикле делить регионы пополам, сужая область поиска
#   пока не найдем нужное число

```python
low = -1000
high = 1000


while low <= high:
    answer = (low + high) // 2
    print(answer)
    voprosik = input(f"Ваше число больше, меньше или равно {answer}?: ")
    if voprosik == "равно":
        print(f'окак так {answer} получается\n ура, наливай')
        break
    elif voprosik == "больше":
        low = answer + 1
        print(low, high)
    elif voprosik == "меньше":
        high = answer - 1
        print(low, high)
    else:
        print("дурачье пиши онли больше меньше или равно")
```
