# Запросить у пользователя - хочет ли он сняться в кино
# Если да - предложить две разных роли - Халк или Локи
#   Если Халк - запросить сколько у него бицепс в объеме 
#   Если бицепс меньше 60 - отправить домой
#   если больше или ровно - он принят
#   Если Локи - спросить кого он больше любит - маму или папу 
#   Если папу - отправить домой
#   Если маму - отправить к папе чтобы спросил кого больше любит 
# если нет - попрощаться 
# во всех условиях обрабатывать белиберду и неправильные выборы ошибкой для пользователя

```python
from rich import print as rprint
```

```python
answer = input("Хочешь сняться в кино? (да/нет) ")
if answer.lower() == "да":
    role = input("У нас крч есть две роли Халк и Локи. На какой стул сядешь? ")
    if role.lower() == "халк":
        bitseps_halk = input("че у тебя по бицепсу ")
        if bitseps_halk.isdigit():
            if int(bitseps_halk) >= 60:
                rprint('[bold green] Принят бро')
            else: 
                rprint('[bold red] домой пошел позорник')
        else:
            rprint('[bold red]цифры вводи дурачье')
    elif role.lower() == "локи":
        family_love = input("Кого больше любишь, маму или папу? ")
        if family_love.lower() == "маму":
            rprint("[bold green] иди к папе и спроси у него кого он больше любит")
        elif family_love.lower() == "папу":
            rprint("[bold red] пошел вон")
        else:
            rprint('[bold yellow]пашел вон атсюда')
    else:
        rprint('[bold yellow]пашел вон атсюда')
else:
    rprint('[bold yellow]пашел вон атсюда')
        
```
